```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import OneHotEncoder
from sklearn.decomposition import PCA
from sklearn.metrics import accuracy_score, roc_auc_score, confusion_matrix, roc_curve
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, VotingClassifier
from sklearn.linear_model import LogisticRegression

# ============================
# LOAD DATA
# ============================

train_path = r"Data/EA_Male_37SNPs_training.csv"
test_path = r"Data/EA_Male_37SNPs_test.csv"

train_df = pd.read_csv(train_path)
test_df = pd.read_csv(test_path)

X_train = train_df.iloc[:, :-1]
y_train = train_df.iloc[:, -1]

X_test = test_df.iloc[:, :-1]
y_test = test_df.iloc[:, -1]

# Convert labels to 0 and 1
y_train = y_train.map({'CTRL': 0, 'CASE': 1})
y_test = y_test.map({'CTRL': 0, 'CASE': 1})

# ============================
# ENCODING
# ============================

encoder = OneHotEncoder(handle_unknown='ignore')

X_train_enc = encoder.fit_transform(X_train).toarray()
X_test_enc = encoder.transform(X_test).toarray()

# ============================
# PCA
# ============================

pca = PCA(n_components=20)
X_train_pca = pca.fit_transform(X_train_enc)
X_test_pca = pca.transform(X_test_enc)

# ============================
# MODELS
# ============================

rf = RandomForestClassifier(n_estimators=100, max_depth=10, n_jobs=-1)
gb = GradientBoostingClassifier()
lr = LogisticRegression(max_iter=500)

ensemble = VotingClassifier(
    estimators=[
        ('rf', rf),
        ('gb', gb),
        ('lr', lr)
    ],
    voting='soft'
)

models = {
    "Random Forest": rf,
    "Gradient Boosting": gb,
    "Logistic Regression": lr,
    "Ensemble (Improved)": ensemble
}

results = {}

# ============================
# FUNCTIONS
# ============================

def plot_conf_matrix(y_true, y_pred, title):
    plt.figure()  # 🔥 FIX: new figure for each confusion matrix
    cm = confusion_matrix(y_true, y_pred)

    plt.imshow(cm, cmap=plt.cm.Blues)
    plt.title(title)
    plt.colorbar()

    plt.xlabel("Predicted")
    plt.ylabel("Actual")

    # Add values inside boxes
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            plt.text(j, i, cm[i, j],
                     ha="center", va="center", color="black")

    plt.show()


def plot_roc(y_true, y_prob, label):
    fpr, tpr, _ = roc_curve(y_true, y_prob)
    plt.plot(fpr, tpr, label=label)

# ============================
# TRAIN + EVALUATE
# ============================

# ROC curve figure (only one)
plt.figure(figsize=(8,6))

for name, model in models.items():
    print(f"\n🔹 Training {name}...")

    model.fit(X_train_pca, y_train)

    y_pred = model.predict(X_test_pca)
    y_prob = model.predict_proba(X_test_pca)[:, 1]

    acc = accuracy_score(y_test, y_pred)
    auc = roc_auc_score(y_test, y_prob)

    results[name] = acc

    print(f"{name} Accuracy: {acc:.4f}")
    print(f"{name} AUC: {auc:.4f}")

    # Confusion Matrix (separate clean plot)
    plot_conf_matrix(y_test, y_pred, f"{name} Confusion Matrix")

    # ROC curve (same plot)
    plot_roc(y_test, y_prob, name)

# Final ROC plot
plt.plot([0, 1], [0, 1], '--')
plt.legend()
plt.title("ROC Curve Comparison")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.show()

# ============================
# ACCURACY COMPARISON
# ============================

plt.figure()
plt.bar(results.keys(), results.values())
plt.title("Model Accuracy Comparison")
plt.xticks(rotation=45)
plt.ylabel("Accuracy")
plt.show()

# ============================
# BEST MODEL
# ============================

best = max(results, key=results.get)

print("\n🏆 BEST MODEL:", best)
print("Accuracy:", results[best])
```
